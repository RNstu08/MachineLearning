```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### 1. Simple Plots with plt.plot() and plt.scatter()

```python
# a) Line Plot
x_line = np.linspace(0, 2 * np.pi, 50) # 50 points from 0 to 2*pi
y_line = np.sin(x_line)

plt.figure(figsize=(8, 4)) # Create a new figure window with a specific size
plt.plot(x_line, y_line)    # Plot y vs x
plt.title("Simple Sine Wave (plt.plot)")
plt.xlabel("Radians")
plt.ylabel("Value")
plt.grid(True)
plt.show() # Display the plot
print("Generated simple line plot.")
```

```python
# b) Scatter Plot
x_scatter = np.random.rand(50) * 10 # 50 random x values
y_scatter = np.random.rand(50) * 10 # 50 random y values
sizes = np.random.rand(50) * 100   # Random sizes for markers
colors = np.random.rand(50)        # Random values for color mapping

plt.figure(figsize=(8, 5))
plt.scatter(x_scatter, y_scatter, s=sizes, c=colors, alpha=0.7, cmap='viridis')
# s: marker size, c: color values, alpha: transparency, cmap: colormap
plt.title("Simple Scatter Plot (plt.scatter)")
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.colorbar(label='Color Intensity') # Add a color bar for the 'c' values
plt.grid(True)
plt.show()
print("Generated simple scatter plot.")
print("-" * 30)
```

#### 2. Plotting from Pandas Objects
- Pandas Series and DataFrames have a .plot() method that uses Matplotlib

```python
print("--- Plotting from Pandas ---")
# Create a sample DataFrame
np.random.seed(1)
dates = pd.date_range('2024-01-01', periods=50)
df = pd.DataFrame(np.random.randn(50, 3).cumsum(axis=0), # Cumulative sum
                  columns=['Value A', 'Value B', 'Value C'],
                  index=dates)

print("Sample Pandas DataFrame (first 5 rows):\n", df.head())
```

```python
# Plotting a DataFrame (plots all columns as lines by default)
df.plot(figsize=(10, 5), title="Plotting DataFrame Columns")
plt.xlabel("Date")
plt.ylabel("Cumulative Value")
plt.grid(True)
plt.legend(loc='upper left') # Adjust legend location
plt.show()
print("\nGenerated line plot directly from DataFrame.")
```

```python
# Plotting a specific column (Series)
df['Value B'].plot(title="Plotting a Single Series (Value B)", legend=True)
plt.show()
print("Generated line plot directly from Series.")
```

```python
# Creating other plot types using df.plot(kind=...)
# Bar plot of the first 5 rows
df.head().plot(kind='bar', title="Bar Plot (First 5 Rows)", rot=0)
plt.ylabel("Value")
plt.tight_layout()
plt.show()
print("Generated bar plot from DataFrame.")
```

```python
# Scatter plot using DataFrame columns
df.plot(kind='scatter', x='Value A', y='Value C', title="Scatter: Value A vs Value C", alpha=0.6)
plt.grid(True)
plt.show()
print("Generated scatter plot from DataFrame.")
print("-" * 30)
```

#### 3. Basic Customization with Pyplot

```python
print("--- Basic Customization ---")
x = np.linspace(0, 5, 20)
y1 = x ** 2
y2 = x ** 1.5
plt.figure(figsize=(8, 5))

# Plotting with color, linestyle, and marker arguments
plt.plot(x, y1,
         color='red',        # Color name or hex code (#FF0000)
         linestyle='--',     # Dashed line ('-', '--', '-.', ':')
         linewidth=2,        # Line thickness
         marker='o',         # Marker style ('o', 's', '^', 'x', '+', ...)
         markersize=5,       # Marker size
         label='y = x^2')    # Label for the legend

plt.plot(x, y2,
         color='blue',
         linestyle=':',      # Dotted line
         linewidth=1.5,
         marker='^',
         markersize=6,
         label='y = x^1.5')

# Adding labels and title
plt.title("Customized Plot Example")
plt.xlabel("X Axis Label")
plt.ylabel("Y Axis Label")

# Setting axis limits
plt.xlim(0, 5) # Set x-axis limits
plt.ylim(0, 30) # Set y-axis limits

# Adding a legend
plt.legend(loc='upper left', fontsize='small') # Control location and font size

# Adding grid lines
plt.grid(True, linestyle='-', alpha=0.5) # Customize grid appearance

plt.show()
print("Generated customized plot.")
print("-" * 30)
```
