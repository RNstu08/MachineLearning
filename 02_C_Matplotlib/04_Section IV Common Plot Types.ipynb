This document provides examples for creating several common plot types using the Matplotlib `Object-Oriented` interface:

* **Line Plots** (`ax.plot`): For showing trends.
* **Scatter Plots** (`ax.scatter`): For relationships between variables.
* **Bar Charts** (`ax.bar`, `ax.barh`): For comparing categories.
* **Histograms** (`ax.hist`): For visualizing distributions.
* **Box Plots** (`ax.boxplot`): For comparing distribution summaries.
* **Pie Charts** (`ax.pie`): For showing proportions.
* **Error Bars** (`ax.errorbar`): For displaying uncertainty.
* **Stem Plots** (`ax.stem`): For discrete sequences.
* **Fill Between** (`ax.fill_between`): For highlighting areas between curves.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd # Used for boxplot example data

# --- Prepare Sample Data ---
np.random.seed(42) # for reproducibility
x = np.linspace(0, 10, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_err = np.random.rand(50) * 0.3 # Error data

x_scatter = np.random.rand(30) * 10
y_scatter = np.random.rand(30) * 10
sizes = np.random.rand(30) * 200
colors = np.random.rand(30)

categories = ['A', 'B', 'C', 'D', 'E']
bar_values = np.random.randint(5, 20, size=len(categories))

hist_data = np.random.randn(1000) # Data for histogram

# Data for boxplot (multiple distributions)
box_data = [np.random.normal(0, std, 100) for std in [1, 2, 0.5]]
box_labels = ['Group 1', 'Group 2', 'Group 3']

pie_sizes = [15, 30, 45, 10]
pie_labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
pie_explode = (0, 0.1, 0, 0) # Explode the 'Hogs' slice

stem_x = np.arange(0, 10)
stem_y = np.sin(stem_x) * 5


# --- Create Plots using Axes Methods ---

# 1. Line Plot (ax.plot)
print("--- 1. Line Plot ---")
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y_sin, label='Sine', color='blue', linestyle='-', marker='o', markersize=3)
ax.plot(x, y_cos, label='Cosine', color='red', linestyle='--', marker='x', markersize=4)
ax.set_title('Line Plot Example')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
ax.grid(True, alpha=0.5)
plt.show()
print("Generated Line Plot.")

# 2. Scatter Plot (ax.scatter)
print("\n--- 2. Scatter Plot ---")
fig, ax = plt.subplots(figsize=(8, 5))
scatter = ax.scatter(x_scatter, y_scatter, s=sizes, c=colors, cmap='viridis', alpha=0.7, label='Random Points')
ax.set_title('Scatter Plot Example')
ax.set_xlabel('X Value')
ax.set_ylabel('Y Value')
ax.grid(True, linestyle=':')
# Add colorbar
cbar = fig.colorbar(scatter)
cbar.set_label('Color Intensity')
plt.show()
print("Generated Scatter Plot.")

# 3. Bar Chart (ax.bar, ax.barh)
print("\n--- 3. Bar Chart ---")
fig, ax = plt.subplots(figsize=(7, 4))
ax.bar(categories, bar_values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'violet'])
ax.set_title('Vertical Bar Chart Example')
ax.set_xlabel('Category')
ax.set_ylabel('Value')
plt.show()
print("Generated Vertical Bar Chart.")

fig, ax = plt.subplots(figsize=(7, 4))
ax.barh(categories, bar_values, color='salmon') # Horizontal bar chart
ax.set_title('Horizontal Bar Chart Example')
ax.set_xlabel('Value')
ax.set_ylabel('Category')
plt.show()
print("Generated Horizontal Bar Chart.")

# 4. Histogram (ax.hist)
print("\n--- 4. Histogram ---")
fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(hist_data, bins=30, color='teal', alpha=0.8, density=False) # density=True normalizes to probability density
ax.set_title('Histogram Example')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
print("Generated Histogram.")

# 5. Box Plot (ax.boxplot)
print("\n--- 5. Box Plot ---")
fig, ax = plt.subplots(figsize=(7, 5))
ax.boxplot(box_data, labels=box_labels, patch_artist=True, # patch_artist=True allows filling boxes
           showmeans=True, # Show the mean marker
           boxprops=dict(facecolor='lightblue', color='blue'),
           medianprops=dict(color='red', linewidth=2),
           meanprops=dict(marker='D', markeredgecolor='black', markerfacecolor='green'),
           whiskerprops=dict(color='blue', linestyle='--'),
           capprops=dict(color='black'))
ax.set_title('Box Plot Example')
ax.set_ylabel('Value')
ax.grid(axis='y', linestyle=':', alpha=0.5)
plt.show()
print("Generated Box Plot.")

# 6. Pie Chart (ax.pie)
print("\n--- 6. Pie Chart ---")
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(pie_sizes, explode=pie_explode, labels=pie_labels, autopct='%1.1f%%',
       shadow=True, startangle=90, colors=['yellowgreen', 'gold', 'lightskyblue', 'lightcoral'])
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax.set_title('Pie Chart Example')
plt.show()
print("Generated Pie Chart.")

# 7. Error Bars (ax.errorbar)
print("\n--- 7. Error Bars ---")
fig, ax = plt.subplots(figsize=(8, 4))
ax.errorbar(x[::5], y_sin[::5], yerr=y_err[::5], fmt='o', color='black', # fmt combines marker/line style
            ecolor='lightgray', elinewidth=3, capsize=5, label='Sine with Error')
ax.set_title('Error Bar Example')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
ax.grid(True, alpha=0.5)
plt.show()
print("Generated Error Bar Plot.")

# 8. Stem Plot (ax.stem)
print("\n--- 8. Stem Plot ---")
fig, ax = plt.subplots(figsize=(8, 4))
markerline, stemlines, baseline = ax.stem(stem_x, stem_y, linefmt='grey', markerfmt='D', bottom=0, label='Discrete Sequence')
plt.setp(markerline, 'markerfacecolor', 'red', 'markeredgecolor', 'black') # Customize markers
plt.setp(stemlines, 'linestyle', ':', 'color', 'darkgrey') # Customize stems
plt.setp(baseline, 'color', 'black', 'linewidth', 1) # Customize baseline
ax.set_title('Stem Plot Example')
ax.set_xlabel('Index')
ax.set_ylabel('Value')
ax.legend()
ax.grid(True, linestyle=':')
plt.show()
print("Generated Stem Plot.")

# 9. Fill Between (ax.fill_between)
print("\n--- 9. Fill Between ---")
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y_sin, label='Sine', color='blue')
# Fill the area between y_sin and 0 where y_sin is positive
ax.fill_between(x, y_sin, 0, where=(y_sin > 0), color='lightblue', alpha=0.5, label='Positive Area')
# Fill the area between y_sin and 0 where y_sin is negative
ax.fill_between(x, y_sin, 0, where=(y_sin < 0), color='lightcoral', alpha=0.5, label='Negative Area')
ax.set_title('Fill Between Example')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
print("Generated Fill Between Plot.")

print("\nPlotting examples complete.")
```
