While `pyplot` is convenient for simple plots, the Object-Oriented (OO) interface provides much more explicit control and is generally preferred for creating more complex figures, especially those with multiple subplots.

The core idea is to work directly with `Figure` and `Axes` objects:

* **`Figure`**: Represents the entire window or canvas that contains all the plot elements.
* **`Axes`**: Represents an individual plot (an area where points/lines/etc. are plotted with axes, labels, titles). A single `Figure` can contain multiple `Axes` objects.

The most common way to start with the OO interface is using `plt.subplots()`.

## Matplotlib: Object-Oriented Interface (Figure & Axes)

*(May 4, 10:26 PM)*

This document explains:

* The concept of `Figure` and `Axes` objects.
* How to create a `Figure` and `Axes` using `plt.subplots()`.
* How to use methods directly on the `Axes` object (`ax.plot`, `ax.set_title`, `ax.set_xlabel`, `ax.grid`, `ax.legend`, etc.) for plotting and customization.
* How to create a grid of subplots and access individual `Axes` objects within the grid.
* The benefits of using the OO interface over `pyplot` for more complex visualizations.

---

From now on, we'll primarily use the Object-Oriented interface in our examples as it's the more robust approach.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# --- 1. Understanding Figure and Axes ---
# Figure: The top-level container for all plot elements.
# Axes: An individual plot within the figure. It contains axis lines, labels, ticks, title, and the plotted data.

# --- 2. Creating Figures and Axes with plt.subplots() ---
# plt.subplots() is the most common way to create a Figure and one or more Axes objects.

print("--- Creating Figure and Axes ---")

# a) Create a Figure and a single Axes object
# fig is the Figure, ax is the Axes
fig, ax = plt.subplots(figsize=(8, 5)) # figsize applies to the Figure

print(f"Figure object: {fig}")
print(f"Axes object: {ax}")

# b) Plotting using Axes methods
x = np.linspace(0, 10, 100)
y = np.sin(x) * np.exp(-x / 5)

ax.plot(x, y, color='purple', label='Damped Sine Wave') # Use ax.plot() instead of plt.plot()

# c) Customizing using Axes methods
ax.set_title("Plot using Object-Oriented Interface") # Use ax.set_title() etc.
ax.set_xlabel("Time (s)")
ax.set_ylabel("Amplitude")
ax.legend()
ax.grid(True, linestyle=':') # Add grid using the Axes method
ax.set_xlim(0, 10) # Set limits using Axes methods
ax.set_ylim(-0.5, 1.0)

# Display the figure
plt.show()
print("Generated plot using OO interface.")
print("-" * 30)


# --- 3. Creating a Figure with Multiple Subplots (Axes) ---
# plt.subplots(nrows, ncols) creates a grid of Axes.

print("--- Multiple Subplots ---")

# Create a figure with a 2x2 grid of Axes
# axs is now a 2D NumPy array of Axes objects
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8), sharex=False, sharey=False)
# sharex/sharey=True makes subplots share the same x/y axis limits and ticks

print(f"Figure object: {fig}")
print(f"Axes array (axs):\n{axs}")
print(f"Shape of Axes array: {axs.shape}")

# Access individual Axes using array indexing (axs[row, col])
ax1 = axs[0, 0] # Top-left subplot
ax2 = axs[0, 1] # Top-right subplot
ax3 = axs[1, 0] # Bottom-left subplot
ax4 = axs[1, 1] # Bottom-right subplot

# Plot different things on each subplot
x_common = np.linspace(0, 2 * np.pi, 50)

# Top-left: Sine wave
ax1.plot(x_common, np.sin(x_common), color='blue')
ax1.set_title('Sine Wave')
ax1.grid(True)

# Top-right: Cosine wave
ax2.plot(x_common, np.cos(x_common), color='red', linestyle='--')
ax2.set_title('Cosine Wave')

# Bottom-left: Scatter plot
ax3.scatter(np.random.rand(20), np.random.rand(20), color='green', marker='^')
ax3.set_title('Scatter Plot')
ax3.set_xlabel('X Random')
ax3.set_ylabel('Y Random')

# Bottom-right: Histogram
ax4.hist(np.random.randn(100), bins=10, color='orange', alpha=0.7)
ax4.set_title('Histogram')
ax4.set_xlabel('Value')
ax4.set_ylabel('Frequency')

# Add an overall title to the Figure
fig.suptitle('Figure with Multiple Subplots (2x2 Grid)', fontsize=16)

# Adjust layout to prevent labels/titles overlapping
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # rect adjusts layout to make room for suptitle

# Display the figure with all subplots
plt.show()
print("Generated figure with 2x2 subplots using OO interface.")
print("-" * 30)


# --- 4. Why Use the Object-Oriented Interface? ---
# - More explicit control over individual plot elements.
# - Easier to manage complex figures with multiple subplots.
# - Code is generally more readable and maintainable for non-trivial plots.
# - Necessary for embedding plots in GUI applications or web frameworks.

print("The OO interface is preferred for complexity, control, and clarity.")
print("-" * 30)

```
