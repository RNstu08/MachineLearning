This section covers plots designed to visualize the relationship between a numerical variable and one or more categorical variables, or to show the distribution within categories. `Seaborn` offers a variety of effective plot types for this purpose.

## Seaborn: Visualizing Categorical Data

This document covers the wide range of plots `Seaborn` offers for categorical data:

* **Categorical Scatter Plots** (`stripplot`, `swarmplot`): Showing individual data points per category.
* **Categorical Distribution Plots** (`boxplot`, `violinplot`, `boxenplot`): Visualizing the distribution summary within each category.
* **Categorical Estimate Plots** (`pointplot`, `barplot`, `countplot`): Showing central tendencies (like mean or count) and confidence intervals/error bars per category.
* **`catplot`** (Figure-level): A versatile function that can create any of the above axes-level plots while also allowing faceting across rows and columns based on other variables.

---

These plots are essential for comparing groups and understanding distributions based on categorical features.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the 'tips' dataset
tips_df = sns.load_dataset("tips")

# --- Basic Inspection ---
print("--- 'tips' Dataset Head ---")
print(tips_df.head())
print("-" * 30)
# Set a consistent theme
sns.set_theme(style="ticks")
```

#### 1. Categorical Scatter Plots (Axes-level)
- Show individual data points distributed by category.

```python
print("--- 1. Categorical Scatter Plots ---")

# a) Strip Plot (sns.stripplot)
# Points are plotted along the categorical axis, with some random jitter added
# on the other axis to reduce overlap (by default).
fig, ax = plt.subplots(figsize=(8, 5))
sns.stripplot(data=tips_df, x="day", y="total_bill", jitter=True, alpha=0.7, ax=ax)
ax.set_title('Strip Plot: Total Bill Distribution by Day')
plt.show()
print("Generated Strip Plot.")
```

```python
# Strip plot with hue
fig, ax = plt.subplots(figsize=(8, 5))
sns.stripplot(data=tips_df, x="day", y="total_bill", hue="smoker", dodge=True, alpha=0.6, ax=ax) # dodge=True separates hues
ax.set_title('Strip Plot: Total Bill by Day (Separated by Smoker)')
plt.show()
print("Generated Strip Plot with Hue.")
print("-" * 20)
```

```python
# b) Swarm Plot (sns.swarmplot)
# Similar to stripplot, but points are adjusted along the categorical axis
# to avoid overlap. Better for visualizing distributions but doesn't scale well to large datasets.
fig, ax = plt.subplots(figsize=(8, 5))
sns.swarmplot(data=tips_df, x="day", y="total_bill", alpha=0.8, size=4, ax=ax) # size controls marker size
ax.set_title('Swarm Plot: Total Bill Distribution by Day')
plt.show()
print("Generated Swarm Plot.")
```

```python
# Swarm plot with hue
fig, ax = plt.subplots(figsize=(9, 6))
sns.swarmplot(data=tips_df, x="day", y="total_bill", hue="sex", dodge=True, size=5, ax=ax)
ax.set_title('Swarm Plot: Total Bill by Day (Separated by Sex)')
plt.show()
print("Generated Swarm Plot with Hue.")
print("-" * 30)
```

#### 2. Categorical Distribution Plots (Axes-level)
- Show abstract representations of distributions within categories.

```python
print("--- 2. Categorical Distribution Plots ---")

# a) Box Plot (sns.boxplot)
# Shows the five-number summary (min, Q1, median, Q3, max) and outliers.
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(data=tips_df, x="day", y="total_bill", ax=ax, palette="Set2")
ax.set_title('Box Plot: Total Bill Distribution by Day')
plt.show()
print("Generated Box Plot.")
```

```python
# Box plot with hue
fig, ax = plt.subplots(figsize=(9, 6))
sns.boxplot(data=tips_df, x="day", y="total_bill", hue="time", ax=ax, palette="husl")
ax.set_title('Box Plot: Total Bill by Day (Separated by Time)')
plt.show()
print("Generated Box Plot with Hue.")
print("-" * 20)
```

```python
# b) Violin Plot (sns.violinplot)
# Combines a box plot with a kernel density estimate (KDE) on each side.
fig, ax = plt.subplots(figsize=(8, 5))
sns.violinplot(data=tips_df, x="day", y="total_bill", ax=ax, palette="muted", inner="quartile") # inner='box', 'quartile', 'point', 'stick', None
ax.set_title('Violin Plot: Total Bill Distribution by Day')
plt.show()
print("Generated Violin Plot.")
```

```python
# Violin plot with hue and split
fig, ax = plt.subplots(figsize=(9, 6))
sns.violinplot(data=tips_df, x="day", y="total_bill", hue="smoker", split=True, inner="quart", ax=ax, palette="coolwarm")
# split=True draws half violins for each hue level
ax.set_title('Violin Plot: Total Bill by Day (Split by Smoker)')
plt.show()
print("Generated Violin Plot with Hue (Split).")
print("-" * 20)
```

```python
# c) Boxen Plot (sns.boxenplot) - Letter-value plot
# Enhanced box plot better suited for larger datasets, showing more quantiles.
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxenplot(data=tips_df, x="day", y="total_bill", ax=ax, palette="pastel")
ax.set_title('Boxen Plot: Total Bill Distribution by Day')
plt.show()
print("Generated Boxen Plot.")
print("-" * 30)
```

#### 3. Categorical Estimate Plots (Axes-level)
- Show central tendency (e.g., mean) and confidence intervals/error bars within categories.

```python
print("--- 3. Categorical Estimate Plots ---")

# a) Point Plot (sns.pointplot)
# Shows point estimates (mean by default) and confidence intervals using vertical lines.
# Good for comparing levels across categories, especially with `hue`. Connects points.
fig, ax = plt.subplots(figsize=(8, 5))
sns.pointplot(data=tips_df, x="day", y="tip", hue="sex", dodge=True, # dodge separates hues slightly
              markers=["o", "s"], linestyles=["-", "--"], ax=ax, palette="Set1")
ax.set_title('Point Plot: Average Tip by Day and Sex')
plt.show()
print("Generated Point Plot.")
print("-" * 20)
```

```python
# b) Bar Plot (sns.barplot)
# Shows point estimates (mean by default) and error bars (confidence interval by default).
fig, ax = plt.subplots(figsize=(8, 5))
sns.barplot(data=tips_df, x="day", y="total_bill", ax=ax, palette="viridis", ci="sd") # ci='sd' shows standard deviation instead of CI
# estimator=np.median # Can change the estimator function
ax.set_title('Bar Plot: Average Total Bill by Day (Error Bars = Std Dev)')
plt.show()
print("Generated Bar Plot.")
```

```python
# Bar plot with hue
fig, ax = plt.subplots(figsize=(9, 6))
sns.barplot(data=tips_df, x="day", y="total_bill", hue="smoker", ax=ax, palette="magma", ci=95) # ci=95 is default bootstrap CI
ax.set_title('Bar Plot: Average Total Bill by Day (Separated by Smoker)')
plt.show()
print("Generated Bar Plot with Hue.")
print("-" * 20)
```

```python
# c) Count Plot (sns.countplot)
# Shows the counts of observations in each category (like a histogram for categorical vars).
fig, ax = plt.subplots(figsize=(8, 5))
sns.countplot(data=tips_df, x="day", ax=ax, palette="rocket")
ax.set_title('Count Plot: Number of Observations per Day')
plt.show()
print("Generated Count Plot.")
```

```python
# Count plot with hue
fig, ax = plt.subplots(figsize=(8, 5))
sns.countplot(data=tips_df, x="day", hue="time", ax=ax, palette="cubehelix")
ax.set_title('Count Plot: Observations per Day (Separated by Time)')
plt.show()
print("Generated Count Plot with Hue.")
print("-" * 30)
```

#### 4. Unified Categorical Plot (sns.catplot - Figure-level)
- Provides a single interface to all the above categorical plots, combined with FacetGrid.

```python
print("--- 4. Unified Categorical Plots (catplot) ---")

# a) Box plot using catplot, faceted by 'time'
sns.catplot(data=tips_df, x="day", y="total_bill",
            col="time",      # Facet columns by time
            kind="box",      # Specify the kind of plot
            height=4, aspect=1.1)
plt.suptitle('Box Plot via catplot (Faceted by Time)', y=1.03)
plt.show()
print("Generated faceted Box Plot using catplot.")
```

```python
# b) Bar plot using catplot, faceted by 'sex'
sns.catplot(data=tips_df, x="day", y="tip", hue="smoker",
            col="sex",       # Facet columns by sex
            kind="bar",      # Specify the kind of plot
            height=4.5, aspect=1)
plt.suptitle('Bar Plot via catplot (Faceted by Sex, Hue by Smoker)', y=1.03)
plt.show()
print("Generated faceted Bar Plot using catplot.")
```

```python
# c) Violin plot using catplot
sns.catplot(data=tips_df, x="day", y="tip", hue="time",
            kind="violin", split=True, height=5, aspect=1.2)
plt.suptitle('Violin Plot via catplot (Split by Time)', y=1.02)
plt.show()
print("Generated Violin Plot using catplot.")
print("-" * 30)
```
