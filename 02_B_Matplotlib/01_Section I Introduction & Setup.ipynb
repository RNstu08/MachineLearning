# What is Matplotlib?

**Purpose:** Matplotlib is the foundational data visualization library in Python. It's used to create a vast range of static, animated, and interactive plots and figures.

**History:** Started by John D. Hunter in 2003, it aimed to provide a plotting capability similar to MATLAB.

**Role:** It forms the base for many other plotting libraries (like Seaborn) and integrates seamlessly with NumPy and Pandas, making it essential for visualizing data analyzed with those tools.

## Architecture (Briefly):

* **Backend Layer:** Handles the actual rendering to different outputs (screen display windows, file formats like PNG, PDF, SVG). Examples: `Agg` (for PNG), `Qt5Agg`, `TkAgg`. Usually, you don't interact with this directly.
* **Artist Layer:** Contains the objects that make up a plot (`Figure`, `Axes`, `Lines`, `Text`, etc.). Most of the work happens here, especially when using the Object-Oriented interface.
* **Scripting Layer (`pyplot`):** A simpler interface (like MATLAB) that automates much of the `Figure` and `Axes` creation. Good for quick plots.

## Interfaces:

* **`pyplot`:** A collection of functions (`plt.plot`, `plt.title`, etc.) that implicitly manage `Figure` and `Axes`. Easier for simple plots but can become confusing for complex ones.
* **Object-Oriented (OO):** Explicitly create and manage `Figure` and `Axes` objects (`fig`, `ax`). You call methods directly on these objects (e.g., `ax.plot()`, `ax.set_title()`). This is more verbose but offers much better control and clarity, especially for figures with multiple subplots. We'll focus more on this interface after introducing the basics.

#### Matplotlib: Basic Setup and Plot

````python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np # Often used for creating data to plot

# --- Basic Plotting with pyplot ---

# 1. Prepare some data
x = np.linspace(0, 10, 100) # 100 points from 0 to 10
y_sin = np.sin(x)
y_cos = np.cos(x)

# 2. Create the plot using pyplot functions
plt.plot(x, y_sin, label='Sine Wave') # Plot sin(x)
plt.plot(x, y_cos, label='Cosine Wave') # Plot cos(x)

# 3. Add basic customization
plt.title('Simple Sine and Cosine Waves')
plt.xlabel('X-axis values')
plt.ylabel('Y-axis values')
plt.legend() # Display the labels defined in plt.plot()
plt.grid(True) # Add a grid

# 4. Display the plot
# In standard Python scripts, plt.show() is essential.
# In Jupyter with %matplotlib inline, it might not be strictly needed for display,
# but it's good practice to signal the plot is complete.
plt.show()

print("Plot generated and displayed.")

# --- Example of creating a figure explicitly (brief preview of OO) ---
# fig, ax = plt.subplots() # Create a figure and a single axes object
# ax.plot(x, y_sin)       # Plot on the axes object
# ax.set_title("Plot using OO interface")
# plt.show()
````
